```python
import matplotlib.pyplot as pl
%matplotlib inline
import numpy as np
import everest

# Download the data
star = everest.Everest(201367065, quiet = True)

# Remove outliers, get a nice subset of the data
cut = range(300,1700)
t = star.apply_mask(star.time)[cut]
F = star.apply_mask(star.fpix)[cut]
```

```python
# Sort the pixels by median flux
inds = np.argsort(np.nanmedian(F, axis = 0))[::-1]
F = F[:,inds]
```

```python
fig = pl.figure(figsize = (18,5))
pl.plot(t, np.sum(F, axis = 1), 'k.-', alpha = 0.35, ms = 3, lw = 1);
```

```python
fig = pl.figure()
pl.imshow(F, aspect = 'auto');
```
